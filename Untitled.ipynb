```python
X = np.random.randint(0, 100, 1000)
Y = X * np.random.randint(38,42, 1000)
```

```python
pos = []
score = []
for i in np.arange(30, 50, 0.1):
    for j in np.arange(-20, 100, 1):
        pred = X*i + j
        score.append(mean_absolute_error(Y, pred))
        pos.append([i, j])
score = np.array(score)
score = 1/score
pos = np.array(pos)
score = score / score.max()
```
