```python
%pylab inline
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from colour import Color
from pylab import rcParams
rcParams['figure.figsize'] = 16, 8
```

```python
data = pd.read_csv("./billes.csv")
```

```python
model = LinearRegression()
X = data[["billes"]].values
Y = data[["poids"]].values
model.fit(X, Y)
```

```python
model.coef_, model.intercept_
```

```python
pos = []
score = []
for i in np.arange(30, 50, 0.1):
    for j in np.arange(-20, 100, 1):
        pred = X*i + j
        score.append(mean_absolute_error(Y, pred))
        pos.append([i, j])
score = np.array(score)
score = 1/score
pos = np.array(pos)
score = score / score.max()
```

```python
colors = list(Color("red").range_to(Color("blue"),100))
scoreColors = [colors[int(i*100)-1].get_hex_l() for i in score]
```

```python
plt.scatter(pos[:,1], pos[:,0], color = scoreColors)
```
